```python
import pandas as pd

# Load the CSV using the full path
df = pd.read_csv(r"C:\Users\empet\Documents\ai-engineering-portfolio\01-data-analysis-sales\sales_data.csv")
df.head()
```

```python
# Check info about the dataset
df.info()

# Quick statistics
df.describe()
```

```python
# Total revenue
df['Revenue'] = df['Units_Sold'] * df['Unit_Price']
print("Total Revenue:", df['Revenue'].sum())

# Average daily revenue
print("Average Revenue per Sale:", df['Revenue'].mean())

# Top selling product by revenue
top_product = df.groupby('Product')['Revenue'].sum().idxmax()
print("Top Selling Product:", top_product)
```

```python
import matplotlib.pyplot as plt

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Plot revenue over time
daily_revenue = df.groupby('Date')['Revenue'].sum()
daily_revenue.plot(figsize=(10,4), marker='o')
plt.title("Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```
